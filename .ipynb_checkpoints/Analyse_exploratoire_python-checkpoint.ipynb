```python
# Exemple de DataFrame fictif pour tester
df = pd.DataFrame({
    'age': [12, 25, 37, 45, 53, 61, 18, 29, 41, 35, 67],
    'sex': ['male', 'female', 'female', 'male', 'female', 'male', 'male', 'female', 'female', 'male', 'male']
})
```

```python
df['age_group'] = pd.cut(df['age'], bins=[0, 10, 20, 30, 40, 50, 60, 100], 
                         labels=['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '60+'])

sexes = df['sex'].unique()

for s in sexes:
    data = df[df['sex'] == s]['age_group'].value_counts().sort_index()
    plt.figure(figsize=(5, 5))
    data.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Répartition des âges ({s})")
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
```

```python
# Créer les tranches d’âge
catdata['age_group'] = pd.cut(
    df['age'],
    bins=[0, 10, 20, 30, 40, 50, 60, 100],
    labels=['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '60+'],
    right=False
)
```

```python
# Afficher un graphique en secteur pour chaque sexe
sexes = df['sex'].dropna().unique()  # ignore les valeurs manquantes

for s in sexes:
    data = df[df['sex'] == s]['age_group'].value_counts().sort_index()
    plt.figure(figsize=(5, 5))
    data.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Répartition des âges ({s})")
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
```

```python
#affichage des ages en graphine histogramme
plt.figure(figsize=(7, 5))
sb.histplot(data=df, x='age', bins=10, kde=True, color='skyblue')
plt.title("Distribution des âges")
plt.xlabel("Âge")
plt.ylabel("Nombre d'individus")
plt.tight_layout()
plt.show()
```
